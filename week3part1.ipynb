```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

# !conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

# !conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')

!conda install -c conda-forge beautifulsoup4 --yes 
# !conda install -c anaconda lxml

from bs4 import BeautifulSoup

source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'html.parser')

row = soup.find('table', class_='wikitable sortable')

ln=[]

for rows in row.find_all('tr'):
    if rows.find_all('td'):
        ln.append(rows.find_all('td'))
        
#change to text the first

i = 0
for n in ln:
    ln[i][0] = n[0].text
    i+=1

second_array = []

i=0
for n in ln:
    if ln[i][1].text != 'Not assigned':
        second_array.append(ln[i])
    i+=1

#change to text in the second

i = 0
for n in second_array:
    second_array[i][1] = second_array[i][1].text
    i+=1

#change to text in the third

i = 0
for n in second_array:
    text = second_array[i][2].text
    second_array[i][2] = text[:-1]
    i+=1

```

```python
second_array
```

```python
#merge and format same values
def mergeSameValue(ls):
    
    merged_array = []
    counter = 0
    neighbor = ''
    i = 0
    
    for n in ls:
        if i+1 < len(ls):
            if ls[i][0] == ls[i+1][0]:
                neighbor += ls[i][2]+', '
                counter+=1
            if ls[i][0] != ls[i+1][0]:
                if counter == 0:
                    row = [ls[i][0], ls[i][1], ls[i][2]]
                    merged_array.append(row)
                if counter != 0:
                    row = [ls[i][0], ls[i][1], neighbor+ls[i+1][2]]
                    merged_array.append(row)
                    neighbor = ''
                    counter = 0
        else:
            row = [ls[i][0], ls[i][1], neighbor+ls[i][2]]
            merged_array.append(row)
                    
        
        i+=1
        
    return merged_array
```

```python
final_array = mergeSameValue(second_array)
```

```python
final_array
```

```python
post_c = []
bor = []
neigh = []

i=0
for n in final_array:
    post_c.append(final_array[i][0])
    bor.append(final_array[i][1])
    neigh.append(final_array[i][2])
    i+=1
```

```python
df = pd.DataFrame(
    {'PostalCode': post_c,
     'Borough': bor,
     'Neighborhood': neigh
    })
```
